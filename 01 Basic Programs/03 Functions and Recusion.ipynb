```python
#binary search

def bs(a,t,s,e):
    
    if s>e:
        return "NOT EXIST"
    
    mid=int((s+e)/2)
    
    if a[mid] == t:
        return mid
    elif t<a[mid]:
        return bs(a,t,s,mid-1)
    elif t>a[mid]:
        return bs(a,t,mid+1,e)
a = [1,2,3,4,5,7,8,9,10,11,12,13,14,15]
t=10

x=bs(a,t,0,len(a)-1)

x
```

```python
def counter(n,c):                                   #counter
    if n<=0:
        return c
    n=int(n//10)
    c=c+1
    return counter(n,c)
    
def reverse(n):                                     #reverse 
    if n==0:
        return 0
    
    last=int(n%10)
    c=counter(n,0)
    n=int(n//10)
    
    
    return (last*(10**(c-1)) + reverse(n))
    
    
        
def pali(n):                                      #palindrome 
    if n == reverse(n):
        return True
    return False
pali(1001)
```

```python
def mergesort(first,second):
    
    i=0
    j=0
    new=[]
    
    while(i<len(first) and  j<len(second)):
        if first[i]<second[j]:
            new.append(first[i])
            i=i+1
        
        else:
            new.append(second[j])
            j=j+1
    
    while i<len(first):
        new.append(first[i])
        i=i+1
        
    while(j<len(second)):
        new.append(second[i])
        j=j+1
    
    return new

def merge(a):
    if len(a)==1:
        return a
    
    
    mid=len(a)//2
    
    left=merge(a[:mid])
    right=merge(a[mid:])
    
    return mergesort(left,right)
    
    
    
a=[9,8,7,6,5,4,3,2,1,0]
x=merge(a)
x
```

```python
def display(board):
    print("")
    for i in board:
        for j in i:
            if j == True:
                print("Q", end=' ')
            else:
                print("X", end=' ')
        print('')


def issafe(board, row, col):
    for i in range(0, row):
        if board[i][col]:
            return False

    maxleft = min(row, col)

    for i in range(1, maxleft + 1):
        if board[row - i][col - i]:
            return False

    maxright = min(row, len(board) - col - 1)

    for i in range(1, maxright + 1):
        if board[row - i][col + i]:
            return False

    return True


def queens(board, row):
    if row == len(board):
        display(board)
        return 1

    count = 0
    for col in range(0, len(board)):
        if issafe(board, row, col):
            board[row][col] = True
            count = count + queens(board, row + 1)
            board[row][col] = False

    return count


board = [[False, False, False, False], [False, False, False, False], [False, False, False, False],
         [False, False, False, False]]

queens(board, 0)
```

```python
def display(board):
    print("")
    for i in board:
        for j in i:
            if j == True:
                print("Q", end=' ')
            else:
                print("X", end=' ')
        print('')


def issafe(board, row, col):
    
    if (-1<row-1<4 and -1<col-2<4 and board[row-1][col-2]):
        return False
    
    if (-1<row-1<4 and -1<col+2<4 and board[row-1][col+2]):
        return False
        
    if (-1<row-2<4 and -1<col-1<4 and board[row-2][col-1]):
        return False
    
    if (-1<row-2<4 and -1<col+1<4 and board[row-2][col+1]):
        return False
    
    return True  

def k(board, row,col,n):
    
    if n==0:
        display(board)
        return 1
    
    if row==len(board)-1 and col==len(board):
        return 0
    
    if col == len(board):
         return k(board, row+1,0,n)
        
    count = 0
    
    if issafe(board, row, col):
        board[row][col] = True
        count += k(board, row,col+1,n-1)
        board[row][col] = False
     

    count+=k(board, row,col+1,n)
    return count



board = [[False, False, False, False], [False, False, False, False], [False, False, False, False],
         [False, False, False, False]]

k(board,0,0,4)
```

```python
from graphviz import Digraph

def visualize_tree(board, row, col, n, graph, parent=None):
    # Node label for the current state
    node_label = f"k({row}, {col}, {n})"
    current_node = f"{row}{col}{n}"
    
    if parent:
        graph.edge(parent, current_node, label=node_label)
    else:
        graph.node(current_node, label=node_label)
    
    if n == 0 or (row == len(board) - 1 and col == len(board)):
        return
    
    if col == len(board):
        visualize_tree(board, row + 1, 0, n, graph, current_node)
    else:
        if issafe(board, row, col):
            board[row][col] = True
            visualize_tree(board, row, col + 1, n - 1, graph, current_node)
            board[row][col] = False
        
        visualize_tree(board, row, col + 1, n, graph, current_node)

# Initialize the graph
graph = Digraph(format='png')

# Create the board
board = [[False, False, False, False] for _ in range(4)]

# Visualize the tree
visualize_tree(board, 0, 0, 2, graph)  # For full n=4, replace 2 with 4
graph.render('knight_tree')
```
