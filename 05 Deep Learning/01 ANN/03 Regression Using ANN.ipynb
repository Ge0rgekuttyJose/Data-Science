### Reading the Dataset

### Removing Duplicates Values

```python
data = data.drop_duplicates()
data
```

```python
data = data.reset_index(drop=True)
data
```

### Converting Categorical Values into Numberical Values

```python
from sklearn.preprocessing import LabelEncoder

names = LabelEncoder()

data['name'] = names.fit_transform(data['name'])
data
```

```python
from sklearn.preprocessing import LabelEncoder

fuels = LabelEncoder()

data['fuel'] = fuels.fit_transform(data['fuel'])
data
```

```python
from sklearn.preprocessing import LabelEncoder

seller_types = LabelEncoder()

data['seller_type'] = seller_types.fit_transform(data['seller_type'])
data
```

```python
from sklearn.preprocessing import LabelEncoder

transmissions = LabelEncoder()

data['transmission'] = transmissions.fit_transform(data['transmission'])
data
```

```python
from sklearn.preprocessing import LabelEncoder

owners = LabelEncoder()

data['owner'] = owners.fit_transform(data['owner'])
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
column_names = ['name', 'year', 'selling_price', 'km_driven', 'fuel', 'seller_type',
       'transmission', 'owner']
```

### Checking for Outlier

```python
i=1

for col in column_names:
    plt.subplots_adjust(wspace=0.6,hspace=0.6)
    plt.subplot(4,4,i)
    i=i+1
    plt.boxplot(data[col],patch_artist=True)
    plt.title(col)
```

```python
data
```

### Removing The Outliers

```python
outlier_columns = ['year', 'selling_price', 'km_driven']

i=1

for col in outlier_columns:
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    
    IQR=Q3-Q1
    
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    
    upper_index=np.where(data[col] >= upper)[0]
    lower_index=np.where(data[col] <= lower)[0]
    
    data.drop(index= upper_index , inplace = True)
    data.drop(index= lower_index , inplace = True)
    
    plt.subplots_adjust(wspace=0.6,hspace=0.6)
    plt.subplot(3,1,i)
    i=i+1
    plt.boxplot(data[col])
    plt.title(col)
    
    data.reset_index(inplace=True)
    data.drop('index',axis=1,inplace=True)
    
    
    
    
    
```

### Spliting into Feature and Target

```python
x=data.drop('selling_price', axis = 1)
```

```python
y=data['selling_price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=120)
```

```python
data
```

### Creating a ANN model

```python
model  = Sequential()

model.add(Dense(50,input_dim = 7,activation= 'relu',kernel_initializer='uniform'))
model.add(Dense(100,activation ='relu',kernel_initializer='uniform'))
model.add(Dense(100,activation= 'relu',kernel_initializer='uniform'))
model.add(Dense(75,activation ='relu'))
model.add(Dense(100,activation ='relu'))
model.add(Dense(100,activation ='relu'))


model.add(Dense(1,activation= 'relu'))

callbacks = tf.keras.callbacks.EarlyStopping(monitor = 'loss',patience = 50)

model.compile(loss= 'mse',optimizer = Adam(learning_rate =0.001))
model.summary()
```

### Traing the model

```python
training_info = model.fit(x_train,y_train,batch_size=200,epochs=100,validation_data=(x_test,y_test),callbacks=callbacks)
```

```python
hist = training_info.history
```

### Training Loss vs Testing Loss

```python
plt.plot(hist['loss'],label='training_loss',color='red')
plt.plot(hist['val_loss'],label='validation_loss',color='green')
plt.xlabel('epoch')
plt.ylabel("loss")
plt.legend()
```

```python
y_pred = model.predict(x_test)
```
