```python
import gensim
from gensim.models import Word2Vec
import matplotlib.pyplot as plt
import numpy as np
```

### Sample Sentence



sentences = [
    ["king","queen",'man','women'],
    ['paris','france','italy'],
    ['dog','cat','animals']

]

### Train Word2vec model 

```python
model = Word2Vec(sentences,vector_size=3,window=1,min_count=1)
```

### Get and Print Embeddings

```python
words = ['king','queen','paris','tree','cat','italy','dog']
embeddings = []
labels =[]
```

```python
for word in words:
    try:
        embedding = model.wv[word]
        print(f"Embedding for {word}:{embedding}")
        embeddings.append(embedding)
        labels.append(word)
    except KeyError:
        print(f"{word} is not in sentences")
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')
embeddings = np.array(embeddings)

x = embeddings[:,0]

y = embeddings[:,1]

z = embeddings[:,2]

ax.scatter(x,y,z)

for i ,label in enumerate(labels):
    ax.text(x[i] ,y[i], z[i], label)
    
ax.set_xlabel("Dimension 1")
ax.set_ylabel("Dimension 2")
ax.set_zlabel("Dimension 3")
ax.set_title("3D Word Embedding")
plt.show()
```
