```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error , mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Datasets/CAR DETAILS.csv')
data
```

### Checking for NULL Values

```python
data.isna().sum()
```

### Checking for Duplicates

```python
data.duplicated().sum()
```

### Removing the Duplicates

```python
data = data.drop_duplicates()
data
```

```python
# Reset the index without modifying the original data reference
data = data.reset_index(drop=True)
data
```

### Converting the Categorical Values into Numerical Values using Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

names = LabelEncoder()

data['name'] = names.fit_transform(data['name'])
data
```

```python
from sklearn.preprocessing import LabelEncoder

fuels = LabelEncoder()

data['fuel'] = fuels.fit_transform(data['fuel'])
data
```

```python
from sklearn.preprocessing import LabelEncoder

seller_types = LabelEncoder()

data['seller_type'] = seller_types.fit_transform(data['seller_type'])
data
```

```python
from sklearn.preprocessing import LabelEncoder

transmissions = LabelEncoder()

data['transmission'] = transmissions.fit_transform(data['transmission'])
data
```

```python
from sklearn.preprocessing import LabelEncoder

owners = LabelEncoder()

data['owner'] = owners.fit_transform(data['owner'])
data
```

```python
data.columns
```

```python
data.dtypes
```

### Checking outliers for Nomial Data

```python
column_names = ['year', 'selling_price', 'km_driven']
```

```python
i=1

for col in column_names:
    plt.subplots_adjust(wspace=0.6,hspace=0.6)
    plt.subplot(4,4,i)
    i=i+1
    plt.boxplot(data[col],patch_artist=True)
    plt.title(col)
```

### Removing outliers for Nomial Data

```python
outlier_columns = ['year', 'selling_price', 'km_driven']

i=1

for col in outlier_columns:
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    
    IQR=Q3-Q1
    
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    
    upper_index=np.where(data[col] >= upper)[0]
    lower_index=np.where(data[col] <= lower)[0]
    
    data.drop(index= upper_index , inplace = True)
    data.drop(index= lower_index , inplace = True)
    
    plt.subplots_adjust(wspace=0.6,hspace=0.6)
    plt.subplot(3,1,i)
    i=i+1
    plt.boxplot(data[col])
    plt.title(col)
    
    data.reset_index(inplace=True)
    data.drop('index',axis=1,inplace=True)
    
    
    
    
    
```

```python
data
```

### Spliting into Independent Variable and Dependent Variable

```python
x=data.drop('selling_price', axis = 1)
```

```python
y=data['selling_price']
```

```python
x
```

```python
y
```

### Spliting the data into training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=120)
```

### Creating the model and training the model

```python
model=LinearRegression()

model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(x_test,y_test)  
```

```python
model.score(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
pd.DataFrame({'True Value' : y_test ,'Predicted Value' : y_pred })
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
