### Support Vector Classifier

```python
import pandas as pd
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

### Reading the Datasets

```python
data = pd.read_csv('waterQuality1.csv')
data
```

### Checking for NULL Values

```python
data.isna().sum()
```

### Checking Data Types

```python
data.dtypes
```

```python
for col in data:
    if data[col].dtype == 'O':
        print('Column name',col)
        print(data[col].unique())
```

```python
data.dtypes
```

### Replacing unwanted Data points

```python
data['ammonia'] = data['ammonia'].replace('#NUM!',0)
data['is_safe'] = data['is_safe'].replace('#NUM!','0')
```

```python
data
```

```python
data.columns
```

### Changing the data point

```python
ob_columns =['ammonia']

for col in ob_columns:
    data[col] = data[col].astype('float')
    data[col] = data[col].replace(0,data[col].mean())
    print(col , data[col].dtype)
```

```python
### Checking for Outliers 
```

```python
import matplotlib.pyplot as plt

column_names = ['aluminium', 'ammonia', 'arsenic', 'barium', 'cadmium', 'chloramine',
       'chromium', 'copper', 'flouride', 'bacteria', 'viruses', 'lead',
       'nitrates', 'nitrites', 'mercury', 'perchlorate', 'radium', 'selenium',
       'silver', 'uranium']


plt.figure(figsize=(20, 35))


for i,col in enumerate(column_names, 1):
    plt.subplots_adjust(wspace=0.5, hspace=0.5)
    plt.subplot(20, 1, i) 
    plt.boxplot(data[col], patch_artist=True)
    plt.title(col)

plt.tight_layout()
plt.show()
data
```

### Spliting into Independent Variable and Dependent Variable

```python
x=data.drop('is_safe', axis = 1)
```

```python
y=data['is_safe']
```

### Spliting the data into training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=120)
```

### Creating the model and training the model

```python
model = SVC()
model.fit(x_train,y_train)
```

### Training vs Testing Score

```python
print("Train Score",model.score(x_train,y_train))
print("Test Score",model.score(x_test,y_test))

```

### Creating the model and training the model

```python
model = SVC(kernel = 'sigmoid')
model.fit(x_train,y_train)
print("Train Score",model.score(x_train,y_train))
print("Test Score",model.score(x_test,y_test))


```

### Creating the model and training the model

```python
model = SVC(kernel = 'linear',C=5,gamma='auto')
model.fit(x_train,y_train)
print("Train Score",model.score(x_train,y_train))
print("Test Score",model.score(x_test,y_test))
```

### Creating the model and training the model

```python
model = SVC(kernel = 'linear',C=5)
model.fit(x_train,y_train)
print("Train Score",model.score(x_train,y_train))
print("Test Score",model.score(x_test,y_test))
```

### Counting target class

```python
from collections import Counter
Counter(y_train)
```

### Using Smoke for equal target in Training

```python
from imblearn.over_sampling import SMOTE

smt = SMOTE()

x_train_smt , y_train_smt = smt.fit_resample(x_train,y_train)
```

```python
from collections import Counter
Counter(y_train_smt)
```

### Modelling with Smoke Data

```python
model = SVC(kernel = 'linear',C=5)
model.fit(x_train,y_train)
print("Train Score",model.score(x_train_smt , y_train_smt))
print("Test Score",model.score(x_test,y_test))
```
