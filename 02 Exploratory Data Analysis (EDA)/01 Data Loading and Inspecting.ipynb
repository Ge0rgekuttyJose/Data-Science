### Graphical Representation of  Data

```python
from matplotlib import pyplot as plt
#import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd

import random
```

### Line Chart 

```python
names = ['a','b','c','d','e']
acutal_weight = [50,60,70,80,10]

plt.figure(figsize=(4,4))

plt.plot(names,acutal_weight)
plt.xlabel("Names")
plt.ylabel("Actual Weight")
plt.title("Weight Diagram")
```

### Multiple Line Chart

```python

acutal_weight = [50,60,70,80,10]
desired_weight=[47,50,64,98,65]

names = ['a','b','c','d','e']

plt.plot(names,acutal_weight,color='green',label='actual_weight')
plt.plot(names,desired_weight,color='red',label='desired_weight')



plt.xlabel("Names")
plt.ylabel("Weight")
plt.title("Multiple Line Plot",{'fontsize':15,'color':'red'})
plt.legend()
```

```python

acutal_weight = [50,60,70,80,10]
desired_weight=[47,50,64,98,65]

names = ['a','b','c','d','e']

plt.plot(names,acutal_weight,color='green',label='actual_weight')
plt.xlabel("Names")
plt.ylabel("Weight")
plt.title("Line Plot of Actual Weight ",{'fontsize':15,'color':'red'})
plt.legend()



plt.figure()


plt.plot(names,desired_weight,color='red',label='desired_weight')



plt.xlabel("Names")
plt.ylabel("Weight")
plt.title("Line Plot of Desired weight",{'fontsize':15,'color':'red'})
plt.legend()
```

### Pie Chart

```python
count=data1.smoking_status.value_counts()
labels=data1.smoking_status.unique()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(count, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Smoking Status")
plt.show()
```
