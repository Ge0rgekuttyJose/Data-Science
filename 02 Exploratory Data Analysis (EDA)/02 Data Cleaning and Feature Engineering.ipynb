### Outliers

```python
column_names = ['name', 'year', 'selling_price', 'km_driven']
```

```python
i=1

for col in column_names:
    plt.subplots_adjust(wspace=0.6,hspace=0.6)
    plt.subplot(4,1,i)
    i=i+1
    plt.boxplot(data[col],patch_artist=True)
    plt.title(col)
    
data
```

```python
outlier_columns = ['name', 'year', 'selling_price', 'km_driven']

i=1

for col in outlier_columns:
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    
    IQR=Q3-Q1
    
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    
    upper_index=np.where(data[col] >= upper)[0]
    lower_index=np.where(data[col] <= lower)[0]
    
    data.drop(index= upper_index , inplace = True)
    data.drop(index= lower_index , inplace = True)
    
    plt.subplots_adjust(wspace=0.6,hspace=0.6)
    plt.subplot(4,1,i)
    i=i+1
    plt.boxplot(data[col])
    plt.title(col)
    
    data.reset_index(inplace=True)
    data.drop('index',axis=1,inplace=True)
    #data.reset_index(drop=True, inplace=True)
    
    
    
    
data
```

### Normalization

Normalization in data science is a technique used to scale data to a standard range, typically between 0 and 1, without changing the relationships or distributions in the data. This process is essential for algorithms that rely on distance metrics, such as K-nearest neighbors (KNN), neural networks, and support vector machines (SVM). There are several normalization techniques commonly used:

Min-Max Normalization: Scales data to a specific range (usually [0,1]) based on the minimum and maximum values of each feature.


```python
from sklearn.preprocessing import MinMaxScaler

scaler= MinMaxScaler()
scaled_data=scaler.fit_transform(data)
pd.DataFrame(scaled_data,columns=data.columns)
```

### -----------------------------------------------------------------------------------------------------------------------------------------------------------

### StandardScaler

Standardization is a data preprocessing technique used to center and scale the data so that it has a mean of zero and a standard deviation of one
